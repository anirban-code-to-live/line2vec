```python
import networkx as nx
from networkx.algorithms.community import greedy_modularity_communities

dataset = 'dolphins'
edgelistPath = '../data/{}/{}.edgelist'.format(dataset,dataset)
G = nx.read_edgelist(edgelistPath)
edges = list(G.edges)

c = list(greedy_modularity_communities(G)) #c contains the node clusters
noClus = len(c)
print 'No of clusters: ' + str(noClus)
edgeLabels = {}
for i in range(len(edges)):
    sourceLabel = [edges[i][0] in l for l in c].index(True)
    destiLable = [edges[i][1] in l for l in c].index(True)
    if sourceLabel == destiLable:
        edgeLabels[edges[i]] = sourceLabel
    else:
        edgeLabels[edges[i]] = -1
```
